```python
# Importing Libraries
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#Data Collection
Keyboard_name=[]
Keyboard_price=[]
Keyboard_rating=[]
#URL of all the pages
for i in range (0,42):
    url='https://www.flipkart.com/search?q=gaming+keyboards&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(i)
    response=requests.get(url)
    soup=BeautifulSoup(response.text)
#converting all the HTML code to readable format using Beautiful Soup
    
    for x in soup.find_all('div',attrs={'class':'_4ddWXP'}):
        kname=x.find('a',attrs={'class':'s1Q9rs'})
        kcost=x.find('div',attrs={'class':'_30jeq3'})
        krating=x.find('div',attrs={'class':'_3LWZlK'})
        
        # for keyboard names
        if kname:
            Keyboard_name.append(kname.text)
        else:
            Keyboard_name.append(np.NaN)
        #for keyboard costs
        if kcost:
            Keyboard_price.append(kcost.text)
        else:
            Keyboard_price.append(np.NaN)
        # for keyboard ratings
        if krating:
            Keyboard_rating.append(krating.text)
        else:
            Keyboard_rating.append(np.NaN)
```

```python
# Creating a Dataframe
df=pd.DataFrame({'Name':Keyboard_name,'Price':Keyboard_price,'Rating':Keyboard_rating})
data=df
df
```

```python
# Converting rating column of string type into a float  type column.
df['Rating']=df['Rating'].astype(float)
```

```python
# Removing currency symbol from price column to convert the column into float type.
for i in range(len(df)):
               df['Price'][i]=df['Price'][i][1:]
```

```python
# Removing commas and converting Price column into float.
df['Price']=df["Price"].str.replace(",","").astype(float)
```
